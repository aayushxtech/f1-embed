```python
import fastf1
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fastf1.Cache.enable_cache('./f1_cache')
```

```python
drivers = ['VER','LEC','SAI']
years = [2022, 2023, 2024]
session_types  = ['FP1', 'FP2', 'FP3', 'Q', 'R']
tracks = ['Silverstone', 'Monaco', 'Spa', 'Suzuka']
all_laps = []
```

```python
for year in years:
    for track in tracks:
        for session_type in session_types:
            try:
                session = fastf1.get_session(year, track, session_type)
                session.load()
                time.sleep(2)

                # Get all laps for selected drivers
                driver_laps = session.laps.pick_drivers(drivers).copy()
                if driver_laps.empty:
                    print(f"No laps found for {year}, Track {track}, {session_type}")
                    continue

                driver_laps['Year'] = year
                driver_laps['SessionType'] = session_type
                driver_laps['Round'] = track
                driver_laps['Track'] = session.event['EventName']

                all_laps.append(driver_laps)

                print(f"Collected {len(driver_laps)} laps from {year} Round {track} {session_type}")

            except Exception as e:
                print(f"Session skipped: {year} Track {track} {session_type} due to error: {e}")
```

```python
weather = session.weather_data.copy()
```

```python
weather.sort_values('Time', inplace=True)
driver_laps.sort_values('LapStartTime', inplace=True)
driver_laps = pd.merge_asof(
    driver_laps, weather,
    left_on='LapStartTime',
    right_on='Time',
    direction='nearest'
)
```

```python
driver_laps.columns.to_list()
```

```python
driver_laps.drop(columns=['Time_x', 'Time_y'], inplace=True)
```

```python
req_columns = [
    'Driver', 'Team', 'LapTime', 'LapNumber', 'Stint',
    'Compound', 'TyreLife', 'FreshTyre',
    'SessionType', 'Year', 'Round', 'Track',
    'AirTemp', 'TrackTemp', 'Humidity', 'Rainfall',
    'Pressure', 'WindSpeed', 'WindDirection'
]
```

```python
columns_to_keep = []

for col in req_columns:
    if col in driver_laps.columns:
        columns_to_keep.append(col)
```

```python
driver_laps = driver_laps[columns_to_keep]
```

```python
driver_laps.columns
```

```python
from fastf1.core import Lap

all_telemetry = []

for year in years:
    for track in tracks:
        for session_type in session_types:
            try:
                session = fastf1.get_session(year, track, session_type)
                session.load()
                print(f"\n✅ Loaded: {year} {track} {session_type}")

                for drv in drivers:
                    try:
                        laps = session.laps.pick_driver(drv).pick_quicklaps()

                        for _, lap in laps.iterrows():
                            try:
                                tel = lap.get_car_data().add_distance()  # ✅ Correct usage

                                tel['Driver'] = drv
                                tel['LapNumber'] = lap['LapNumber']
                                tel['Year'] = year
                                tel['Track'] = track
                                tel['SessionType'] = session_type

                                all_telemetry.append(tel)

                            except Exception as e:
                                print(f"❌ Telemetry failed for {drv} Lap {lap['LapNumber']}: {e}")
                    except Exception as e:
                        print(f"⚠️ Failed to get laps for driver {drv} in {year} {track} {session_type}: {e}")
            except Exception as e:
                print(f"⚠️ Session skipped: {year} {track} {session_type} due to: {e}")
```

```python
all_telemetry
```

```python
telemetry_df = pd.concat(all_telemetry, ignore_index=True)
```

```python
combined_df.head()
```

```python
telemetry_df['Track'] = telemetry_df['Track'].str.strip().str.lower()
combined_df['Track'] = combined_df['Track'].str.strip().str.lower()
```

```python
print("Telemetry Tracks:", telemetry_df['Track'].unique())
print("Combined Tracks:", combined_df['Track'].unique())
```

```python
merge_keys = ['Driver', 'LapNumber', 'Year', 'SessionType']

for df in [telemetry_df, combined_df]:
    df['Driver'] = df['Driver'].astype(str).str.strip().str.upper()
    df['LapNumber'] = df['LapNumber'].astype(str).str.strip()
    df['Year'] = df['Year'].astype(str).str.strip()
    df['SessionType'] = df['SessionType'].astype(str).str.strip().str.upper()

telemetry_df['Track'] = telemetry_df['Track'].str.strip().str.lower()
combined_df['Round'] = combined_df['Round'].str.strip().str.lower()

telemetry_df.rename(columns={'Track': 'Round'}, inplace=True)
merge_keys.append('Round')

# Final merge
df = pd.merge(telemetry_df, combined_df, on=merge_keys, how='inner')

print("Final Merged DataFrame shape:", df.shape)
```
